```python
import pandas as pd
```

```python
dane = pd.read_parquet("dane.parquet")
```

```python
dane = dane.sort_values(["id_drogerii", "data"]).reset_index(drop=True)
```

```python
dane.rename(columns={"data" : "data_sprzedazy"}, inplace=True)
dane["data_odciecia"] = dane["data_sprzedazy"] - pd.DateOffset(days=7)
dane["srednia_3d"] = dane.sprzedaz.shift(7).rolling(window=7, min_periods=3).mean()
dane = dane[["data_odciecia", "data_sprzedazy", "id_drogerii", "sprzedaz", "cena_ropy", "miasto", "wojewodztwo", "typ_drogerii", "srednia_3d"]]
dane.head(15)
```

```python
dane_2 = dane.query("id_drogerii==2")
dane_2.head(15)
```

```python
mazowieckie = dane[dane["wojewodztwo"] == "Mazowieckie"]
pomorskie = dane[dane.wojewodztwo == "Pomorskie"]
```

```python
mazowieckie.sprzedaz.agg(["mean", "median"])
```
